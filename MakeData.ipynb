```python
import pandas as pd

df = pd.read_csv('Hotel Reservations.csv')


for clss in df['booking_status'].unique():
    print(len(df[df['booking_status'] == clss]))


pollute_ratio = 0.05

change = df.sample(int(pollute_ratio*len(df))).index
df.loc[change,'type_of_meal_plan'] = None

pollute_ratio = 0.01

change = df.sample(int(pollute_ratio*len(df))).index
df.loc[change,'no_of_previous_cancellations'] = None


pollute_ratio = 0.05

change = df.sample(int(pollute_ratio*len(df))).index
df.loc[change,'required_car_parking_space']*=1000


df['family_size']  = df['no_of_children']+5*df['no_of_adults']

print(df[['family_size','no_of_adults']].corr())

column_to_move = df.pop("booking_status")
df.insert(len(df.columns), "booking_status", column_to_move)


df.info()
df.describe()

df.to_csv('Hotel Reservations_dtst.csv')
```
